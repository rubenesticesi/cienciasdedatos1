# 2. Analysis Univariado en Python
**El análisis estadístico univariado** en ciencia de datos es el análisis de una única variable a la vez para describir sus características principales, distribuciones de frecuencia y patrones. A diferencia de otros análisis, se enfoca en entender cada variable de forma aislada, sin investigar sus relaciones o causas con otras variables. Se utilizan herramientas como la media, mediana, moda, desviación estándar, histogramas y tablas de frecuencia para resumir y describir los datos.

**Objetivos principales**

**-Describir los datos**: El objetivo principal es resumir y describir las características de una sola variable.

**-Identificar patrones**: Ayuda a descubrir patrones dentro de los datos de esa variable.

**-Comprender la distribución**: Revela cómo se distribuyen los valores de la variable.
¿Cuándo se utiliza?
Cuando necesitas entender las características individuales de cada variable en un conjunto de datos.
Es la forma más simple de análisis de datos.

## Código Python que calcula las estadísticas descriptivas para la variable A9_Score
incluyendo media, mediana, moda, desviación estándar y detección de valores atípicos usando el método IQR:

````python
# GENERAR ESTADISTICAS DESCRIPTIVAS PARA LA VARIABLE ESCOGIDA

import pandas as pd

# Cargar datos
df = pd.read_csv('autism_screening.csv')

# Variable a analizar
var = 'A9_Score'

# Calcular estadísticas descriptivas
media = df[var].mean()
mediana = df[var].median()
moda = df[var].mode().iloc[0] if not df[var].mode().empty else None
std_dev = df[var].std()

# Detectar valores atípicos usando el método de IQR
Q1 = df[var].quantile(0.25)
Q3 = df[var].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)][var]

# Resultados
resultados = {
    'media': media,
    'mediana': mediana,
    'moda': moda,
    'desviacion_estandar': std_dev,
    'limite_inferior_outliers': lower_bound,
    'limite_superior_outliers': upper_bound,
    'valores_atipicos': outliers.tolist()
}

print(resultados)
````

Esto indica que la mayoría de los valores de A9_Score están concentrados cerca de cero, con poca desviación y sin valores atípicos detectados según el método IQR

````python
# GENERAR ESTADISTICAS DESCRIPTIVAS PARA TODAS VARIABLES DEL DATASET

# Seleccionar solo columnas numéricas
num_cols = ['A1_Score','A2_Score','A3_Score','A4_Score','A5_Score','A6_Score','A7_Score','A8_Score','A9_Score','A10_Score','age','result']

# Calcular estadísticas descriptivas
desc_stats = df[num_cols].describe()
print("Descripción Básica:\n", desc_stats)

# Calcular moda
for col in num_cols:
    print(f"Moda para {col}: ", df[col].mode().values)

# Calcular medida de dispersión
for col in num_cols:
    print(f"Desviación estándar de {col}:", df[col].std())

# Calcular mediana
for col in num_cols:
    print(f"Mediana de {col}:", df[col].median())

# Detección de valores atípicos usando IQR
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    print(f"Valores atípicos por IQR en {col}: {outliers.shape} registros")

# Detección de valores atípicos con z-score
from scipy import stats
for col in num_cols:
    z_scores = np.abs(stats.zscore(df[col].dropna()))
    outliers = np.where(z_scores > 3)
    print(f"Valores atípicos por Z-score en {col}: {len(outliers[0])} registros")
````
