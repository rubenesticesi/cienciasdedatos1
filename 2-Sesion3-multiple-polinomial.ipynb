###  **Extensi贸n al Modelo de Regresi贸n Lineal M煤ltiple**


La **regresi贸n lineal m煤ltiple** es una extensi贸n de la regresi贸n lineal simple que permite modelar la relaci贸n entre una variable dependiente $ y $ y m煤ltiples variables independientes $ x_1, x_2, \dots, x_n $. 
Este modelo es ampliamente utilizado en estad铆stica y aprendizaje autom谩tico para predecir valores num茅ricos y analizar la relaci贸n entre m煤ltiples variables.
En el modelo de regresi贸n lineal m煤ltiple, la relaci贸n se extiende a m煤ltiples variables independientes. Matem谩ticamente, el modelo se expresa como:


$ h(\mathbf{x}) = \theta_0 + \theta_1 \cdot x_1 + \theta_2 \cdot x_2 + \ldots + \theta_n \cdot x_n $

donde:
- $ h(\mathbf{x}) $: valor estimado de la variable dependiente.
- $ \theta_0 $: intercepto.
- $ \theta_1, \theta_2, \ldots, \theta_n $: coeficientes de las variables independientes $ x_1, x_2, \ldots, x_n $.

Los supuestos mencionados anteriormente se mantienen en el modelo m煤ltiple, con la consideraci贸n adicional de que no debe existir multicolinealidad perfecta entre las variables independientes, es decir, las variables explicativas no deben estar perfectamente correlacionadas entre s铆.

En notaci贸n matricial, la ecuaci贸n se puede reescribir como:  

$
 \mathbf{y} = h(\mathbf{x}) = \mathbf{X} \boldsymbol{\theta} + \boldsymbol{\varepsilon}
$

donde:  
- $ \mathbf{y} $ es un vector columna de dimensi贸n $ m \times 1 $ con las observaciones de la variable dependiente,  
- $ \mathbf{X} $ es una matriz de dimensi贸n $ m \times (n+1)$) que contiene las observaciones de las variables independientes, donde la primera columna es un vector de unos para incluir el t茅rmino $ \theta_0 $,  
- $ \boldsymbol{\theta} $ es un vector de dimensi贸n $ (n+1) \times 1 $ que agrupa los par谩metros del modelo,  
- $ \boldsymbol{\varepsilon} $ es un vector de errores de dimensi贸n $ m \times 1 $.  

El objetivo del modelo es encontrar los valores de $ \boldsymbol{\theta} $ que minimicen la funci贸n de costo, generalmente definida como el **error cuadr谩tico medio (MSE)**:

$
J(\boldsymbol{\theta}) = \frac{1}{2m} \sum_{i=1}^{m} (h_{\boldsymbol{\theta}}(x^{(i)}) - y^{(i)})^2
$


Para minimizar $ J(\boldsymbol{\theta}) $, se suelen emplear m茅todos como la **inversa de la matriz normal** (cuando $ \mathbf{X}^T\mathbf{X} $ es invertible) o algoritmos iterativos como el **descenso de gradiente**.

#### **Multicolinealidad en Regresi贸n Lineal M煤ltiple**

Generamos datos donde dos variables independientes est谩n altamente correlacionadas:

```python
m = 100
x = np.linspace(0, 10, m)  # m puntos equidistantes entre 0 y 10
# Generaci贸n de datos con multicolinealidad
x2 = x + np.random.normal(0, 0.01, m)  # Alta correlaci贸n con x
y_multi = 2 + 3 * x + 1.5 * x2 + np.random.normal(0, 1, m)

# Creaci贸n del DataFrame
data_multi = pd.DataFrame({'x1': x, 'x2': x2, 'y': y_multi})

# Ajuste del modelo m煤ltiple
X_multi = sm.add_constant(data_multi[['x1', 'x2']])  # Agrega el intercepto
model_multi = sm.OLS(data_multi['y'], X_multi).fit()

# Resumen del modelo
print(model_multi.summary())
```

```python
# Matriz de correlaci贸n
corr_matrix = data_multi[['x1', 'x2']].corr()
print("Matriz de correlaci贸n:")
print(corr_matrix)
```

#### **Interpretaci贸n**:
1. **Coeficientes no significativos**: 
   - Aunque las variables son predictoras importantes, la multicolinealidad reduce la precisi贸n de las estimaciones de los coeficientes.
2. **Matriz de correlaci贸n**: 
   - Si la correlaci贸n entre las variables independientes es cercana a 卤1, existe multicolinealidad.

### Estimar la relevancia de los coeficientes  

Ahora que tienes los coeficientes, 驴c贸mo puedes determinar si son relevantes para predecir tu variable objetivo?  

La mejor manera es calcular el **p-valor**. El p-valor se utiliza para cuantificar la **significancia estad铆stica**; permite determinar si se debe rechazar o no la hip贸tesis nula.  

### 驴Qu茅 es la hip贸tesis nula?  

En cualquier tarea de modelado, la hip贸tesis de trabajo supone que existe una correlaci贸n entre las variables independientes y la variable objetivo. La **hip贸tesis nula**, en cambio, establece lo contrario: **no hay correlaci贸n** entre las variables independientes y la variable objetivo.  

Por lo tanto, calcular el p-valor para cada coeficiente permite determinar si la variable es estad铆sticamente significativa para predecir el objetivo. Como regla general, si el **p-valor es menor a 0.05**, existe una relaci贸n fuerte entre la variable y la variable objetivo.

###  **Regresi贸n Polin贸mica**

La **regresi贸n polin贸mica** es una extensi贸n de la regresi贸n lineal que permite modelar relaciones no lineales entre las variables. En la regresi贸n lineal simple, la relaci贸n entre la variable dependiente $ y $ y la variable independiente $ x $ se expresa como:

$
h(x) = \theta_0 + \theta_1 x
$

Sin embargo, en muchos casos, los datos pueden seguir un patr贸n no lineal. Para capturar esta relaci贸n, podemos extender la ecuaci贸n a una funci贸n polin贸mica de grado \( n \):

$
h(x) = \theta_0 + \theta_1 x + \theta_2 x^2 + \theta_3 x^3 + \dots + \theta_n x^n
$

Aqu铆, los coeficientes $ \theta_0, \theta_1, ..., \theta_n $ son los par谩metros del modelo que se ajustan mediante el algoritmo de regresi贸n. La regresi贸n polin贸mica sigue siendo un modelo **lineal en los par谩metros**, pero la variable independiente se eleva a diferentes potencias.

###  **Ejemplo en Python: Comparaci贸n entre regresi贸n lineal y polin贸mica**
- **La regresi贸n lineal** no captura la verdadera forma de los datos.  
------
 

```python
np.random.seed(42)
x = np.linspace(-3, 3, 30).reshape(-1, 1)
y = x**3 - 2*x**2 + x + np.random.randn(30, 1) * 2  # Relaci贸n c煤bica con ruido

# Agregar t茅rmino de sesgo para regresi贸n lineal
X_lin = sm.add_constant(x)
# Ajustar modelo de regresi贸n lineal con statsmodels
lin_model = sm.OLS(y, X_lin).fit()
y_pred_lin_sm = lin_model.predict(X_lin)

# Datos generados
plt.scatter(x, y, color='blue', label='Datos generados')
# Regresi贸n Lineal (statsmodels)
plt.plot(x, y_pred_lin_sm, color='red', linestyle='--', label='Regresi贸n Lineal')

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

###  Crear variables polin贸micas manualmente (grado 3)

```python
X_poly = np.column_stack((x, x**2, x**3))
X_poly = sm.add_constant(X_poly)  # Agregar t茅rmino de sesgo
# Ajustar modelo de regresi贸n polin贸mica con statsmodels
poly_model = sm.OLS(y, X_poly).fit()
y_pred_poly_sm = poly_model.predict(X_poly)
```

```python
# Datos generados
plt.scatter(x, y, color='blue', label='Datos generados')
# Regresi贸n Polin贸mica (statsmodels, grado 3)
plt.plot(x, y_pred_poly_sm, color='green', label='Regresi贸n Polin贸mica (grado 3)')

# Regresi贸n Lineal (statsmodels)
plt.plot(x, y_pred_lin_sm, color='red', linestyle='--', label='Regresi贸n Lineal')

plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Comparaci贸n entre Regresi贸n Lineal y Polin贸mica con Statsmodels")
plt.show()
```


En la gr谩fica puedes ver la diferencia entre la regresi贸n lineal (l铆nea roja discontinua) y la regresi贸n polin贸mica (l铆nea verde).  

- **La regresi贸n lineal** no captura la verdadera forma de los datos, ya que asume una relaci贸n estrictamente lineal.  
- **La regresi贸n polin贸mica (grado 3)** logra adaptarse mejor a la tendencia real de los datos, que sigue una relaci贸n c煤bica.
